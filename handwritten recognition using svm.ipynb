```python
X = data.drop('label',axis=1)
y = data['label']
```

```python
#SPLITING THE DATASET INTO TRAIN AND TEST
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
y.shape
```

```python
data.shape
```

```python
cls = SVC(gamma = 'scale')
```

```python
cls.fit(X_train,y_train)
```

```python
#USING THE SVC CLASSIFIER
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
cls.score(X_train,y_train)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
parameters = {'kernel':('linear', 'rbf'), 'C':[1,10,100,1000,10000]}
```

```python
gs = GridSearchCV(cls,parameters)
gs
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_estimator_
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
predicton = gs.predict(X_test)
```

```python
print(classification_report(y_test,predicton))
```

```python
print(confusion_matrix(y_test,predicton))
```
